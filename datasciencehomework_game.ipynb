### MATPLOTLIB
Matplot, verileri çizmemize yardımcı olan bir python kütüphanesidir. En kolay ve en temel grafikler line, scatter ve histogram grafikleri.
* Line plot, x ekseni zaman olduğunda daha iyidir.

### EXPLORATORY DATA ANALYSIS
value_counts(): Frekans sayıları
<br>outliers: verilerin geri kalanından önemli ölçüde daha yüksek veya daha düşük olan değer
* Diyelim ki %75'teki değer Q3 ve %25'teki değer Q1.
* Outlier, Q1 - 1.5(Q3-Q1)'den küçük ve Q3 + 1.5(Q3-Q1)'den daha büyük. (Ç3-Q1) = IQR
<br>describe() method kullanacağız. Describe method içerir:
* count: girdilerin sayısı
* mean: girişlerin ortalaması
* std: standart sapma
* min: minimum giriş
* 25%: ilk quantile
* 50%: medyan veya ikinci quantile
* 75%: üçüncüquantile
* max: maximum giriş

<br>Quantile ne ?

* 1,4,5,6,8,9,11,12,13,14,15,16,17
* Medyan, dizinin **ortasında** bulunan sayıdır. Bu durumda 11 olur.
* Alt çeyrek, en küçük sayı ile medyan arasındaki medyandır, yani 1 ile 11 arasında, yani 6'dır.
* Üst çeyrek, medyan ile en büyük sayı arasındaki medyanı, yani yukarıdaki soruya göre 14 olacak olan 11 ile 17 arasında bulursunuz.
